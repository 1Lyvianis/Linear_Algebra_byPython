```python
###########################check independence by the rref method previously used in project 1
#note that B = Aᵀ
def check_linearly_independent(B,row,col):
    B = copy.deepcopy(B)#to prevent call by reference, we copy the value inside B to a new memory
    for i in range(row):
        if B[i,i]==0:#check first since pivots should not be zero
            for k in range(row):
                if B[k,i]!=0:#search for the first nonzero value
                    for j in range(col):
                        B[i,j],B[k,j]=B[k,j],B[i,j]
                    #Use multiple assignment to swap the element at each index in the array.
                    break  #pivot is now guaranteed nonezero
        pivot = B[i,i]
        for j in range(col):
            B[i,j] /= pivot #each element on the same row is divided by the pivot
        for k in range(row):
            if B[k,i]==0 or k==i:#skip if the element on the same column is zero or itself
                continue
            quo = B[k,i]#set the quotient as main eliminating target, which are the elements below and above pivot
            for j in range(col):
                B[k,j]-=quo * B[i,j]#elements on the same ROW are eliminated

    for i in range(row):
        if B[i,i] != 1:#if full-rank check fails
            print("InputError : Matrix A must be linearly independent columns!")
            return 0
    return 1
#clear screen if independent check fails, reference: https://www.tutorialspoint.com/how-to-clear-screen-in-python
def screen_clear():
   # for mac and linux(here, os.name is 'posix')
   if os.name == 'posix':
      _ = os.system('clear')
   else:
      # for windows platfrom, os.name can be 'nt'
      _ = os.system('cls')
       
#start of main function    
import numpy as np
import copy
import os
#loop until the input matrix A is column-wise linearly independent
while(1):
    print("Input the dimensions of matrix A :")
    row,col = map(int,input().split())
    if(row<col):
        screen_clear()
        print("InputError : Impossible to be linearly independent in an Underdetermined system!")
        continue
    print("\n")
    A = np.zeros((row,col))
    print("Input matrix A :")
    for i in range(row):
        temp = input().split()#split the input numbers, and store them in a temporary array
        for j in range(col):
            A[i,j] = temp[j]#insert the splitted numbers in the main array
    print("\n")        
    #start examining independence. Transpose matrix A as B better computing convenience
    B = A.T # B = Aᵀ
    if(check_linearly_independent(B,col,row)):#independence check passed
        print("Column vectors of A :")
        for i in range(col):
            print("u%d"%(i+1),"=",B[i])#perform placeholder in python
        break
print("\n")
###########################start of Gram-Schmidt process
#a quick swap for better regonition, as we are calculating using matrix B. Note that B = Aᵀ
temp = row
row = col
col = temp
#orgo stands for ORthoGOnal
#we do the manipulation by deepcopying B
orgo = copy.deepcopy(B)
#parameter i represents index of the array
for i in range(1,row):
    for cnt in range(i):#parameter cnt stands for how many projections shall we perform
        squared_length = 0 # based on the formula of projection, we need the length powered by two
        for j in range (col):#parameter j is used while fetching the length of row vector
            squared_length += orgo[cnt, j] * orgo[cnt, j]            
        orgo[i] -= (B[i].dot(orgo[cnt]) / squared_length) * orgo[cnt]#the formula of vector projection

print("Orthogonal basis :")
for i in range(row):
    print("v%d"%(i+1),"=",orgo[i])#perform placeholder in python
print("\n")
###########################start of vector unitization
#orno stands for ORthoNOrmal
orno = copy.deepcopy(orgo)
import math#for square root
#create an 1-dim array to store the length of a row
length = np.zeros((row))
for i in range(row):
    for j in range(col):
        length[i] += orgo[i,j]*orgo[i,j]#accumulating
    for j in range(col):
        orno[i,j] /= math.sqrt(length[i])#deviding
print("Orthonormal basis :")
for i in range(row):
    print("u%d"%(i+1),"=",orno[i])#perform placeholder in python
print("\n")
###########################start of QR-decomposition
print("QR :")
Q = orno.T
R = orno.dot(A)#  Qᵀ * A = Qᵀ * Q * R = R, since Q is an orthogonal matrix
print("Q = \n",Q)
print("R = \n",R)
print("\n")
print("A = \n",Q.dot(R))
```
